```python
import pandas as pd

# Import first MyCNN ternary classification results
myCNN1_df = pd.read_csv("experiments.csv")
```

```python
# Inspect columns
myCNN1_df[["acc_normal","acc_benign","acc_malignant"]]
```

```python
# Get averages of three experiments
cols_to_average = ["acc_normal","acc_benign","acc_malignant"]

myCNN1_avg_df = (
    myCNN1_df.iloc[:-1][cols_to_average]
      .groupby(myCNN1_df.iloc[:-1].index // 3)
      .mean()
      .reset_index(drop=True)
)
```

```python
# Display averages
myCNN1_avg_df
```

```python
# Import second MyCNN ternary classification results
myCNN2_df = pd.read_csv("experiments_v2.csv")
```

```python
# Total time in hours and average time in minutes taken
print((myCNN2_df["training_time_min"].sum())/60)
print(myCNN2_df["training_time_min"].mean())
```

```python
# Get averages of three experiments
cols_to_average2 = ["training_time_min","acc_normal","acc_benign","acc_malignant"]

myCNN2_avg_df = (
    myCNN2_df[cols_to_average2]      # select columns
      .groupby(myCNN2_df.index // 3)
      .mean()
      .reset_index(drop=True)
)
```

```python
# Display averages
myCNN2_avg_df
```

```python
# Create a column with the sum of accuracies
myCNN2_avg_df["acc_total"] = myCNN2_avg_df["acc_normal"] + myCNN2_avg_df["acc_benign"] + myCNN2_avg_df["acc_malignant"]
```

```python
# Display results
myCNN2_avg_df
```

```python
# Weighted residuals calculated as (1 - accuracy) multiplied by an importance factor (normal: 1, benign: 2, malignant: 4)
cols = ["acc_normal", "acc_benign", "acc_malignant"]

weights = pd.Series({
    "acc_normal": 1,
    "acc_benign": 2,
    "acc_malignant": 4
})

residuals = 1 - myCNN2_avg_df[cols]

# myCNN2_avg_df["residual_sum"] = residuals.sum(axis=1)
myCNN2_avg_df["weighted_residual"] = (residuals * weights).sum(axis=1)
```

```python
# Display
myCNN2_avg_df
```

```python
# Import ResNet50 ternary classification results 
ternary_resnet50_df = pd.read_csv("experiments/resnet50_v1/experiments.csv")
```

```python
# Total accuracies
ternary_resnet50_df["acc_total"] = ternary_resnet50_df["acc_normal"] + ternary_resnet50_df["acc_benign"] + ternary_resnet50_df["acc_malignant"]

# Residuals
residuals = 1 - ternary_resnet50_df[cols]
ternary_resnet50_df["weighted_residual"] = (residuals * weights).sum(axis=1)

# Display
ternary_resnet50_df[["training_time_min","acc_normal","acc_benign","acc_malignant", "acc_total", "weighted_residual"]]
```

```python
# Total time in hours and average time in minutes taken
print((ternary_resnet50_df["training_time_min"].sum())/60)
print(ternary_resnet50_df["training_time_min"].mean())
```

```python
# Import ResNet50 binary classification results
binary_resnet50_df = pd.read_csv("experiments/binary_resnet50_roc_auc/roc_auc_binary_experiments.csv")
```

```python
# Total time in hours and average time in minutes taken
print((binary_resnet50_df["training_time_min"].sum())/60)
print(binary_resnet50_df["training_time_min"].mean())
```

```python
# Total accuracies
binary_resnet50_df["acc_total"] = binary_resnet50_df["acc_normal"] + binary_resnet50_df["acc_abnormal"]

# Residuals
residuals = 1 - binary_resnet50_df[["acc_normal","acc_abnormal"]]
binary_weights = pd.Series({
    "acc_normal": 1,
    "acc_abnormal": 3
})
binary_resnet50_df["weighted_residual"] = (residuals * binary_weights).sum(axis=1)

# Display
binary_resnet50_df[["training_time_min","acc_normal","acc_abnormal","acc_total","weighted_residual"]]
```
